## Pulling Data

```python
import requests 
from pattern import web 
# from BeautifulSoup import BeautifulSoup 
```

Find and print the movie title, list of genres, runtime, and score of all movies on this [page](http://www.imdb.com/search/title?at=0&sort=num_votes,desc&start=1&title_type=feature&year=1950,2012). 

<p> <p/> 

```python
url = 'http://www.imdb.com/search/title?sort=num_votes,desc&start=1&title_type=feature&year=1950,2012'
r = requests.get(url)
print r.url 
```

```python
website_html = requests.get(url).text
```

```python
dom = web.Element(website_html)
print dom
```

<br> <br/>

## Picking Data with Pattern

```python
for item in dom.by_tag("div.lister-item-content"):
    movie_name = item.by_tag('h3')[0].by_tag('a')[0].content         
    rating = item.by_tag('strong')[0].content
    year = item.by_tag('span.lister-item-year')[0].content
    year = year.replace("(", "").replace(")", "")
    genre = item.by_tag('span.genre')[0].content
    director = item.by_tag('p')[2]('a')[0].content
    stars = [star.content for star in item.by_tag('p')[2]('a')[1:]]
    movie_description = item.by_tag('p.text-muted')[1].content
    print "Movie Name:", movie_name
    print "Rating:", rating
    print "Year:", year
    print "Genre:", genre 
    print "Director:", director
    print "Stars:", 
    for i in range(len(stars)): 
        print stars[i]
    print "Description:", movie_description
    print "" 
    print ""
    print ""
 
```
